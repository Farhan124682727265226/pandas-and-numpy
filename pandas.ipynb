```python
import pandas as pd
import numpy as np

a=np.arange(4,10)

s=pd.Series(a)

print(s)
# print(s.ndim)
# print(s.size)
# print(s.shape)

print(s.head()) # intial row

print(s.tail()) # last row by defualt get 5 row
```

```python
#create two series

a=pd.Series([4,5,6,7,8])
b=pd.Series([5,4,3,6,7])

print(a)
print(b)

d=pd.concat([a,b],ignore_index=True) #concat - combine two series
print(d)
```

```python
#series addition

# e=a.add(b)
# print(e)

# q=a.sub(b)
# print(q)

w=a.mul(b)
print(w)

# r=a.div(b)
# print(r)
```

```python
z=pd.Series([3,4,5,6],index=["a","b","c","d"]) # to get custom index
print(z)
```

```python
#dictionary

d={"id":1,"age":66,"class":"ds"}
z=pd.Series(d)
print(z)
```

```python
z=pd.Series(4,index=[1,2,3,4])
z
```

```python
z=pd.Series([1,2,3,4,5,6,7])
print(z)
z[3]
print(z[1:3])
z[[6,1,5]]   #to get multiple index [[]]
```

```python
#data frame (row column)

#creating dat frame from list

l=[1,2,3,4,5]

df=pd.DataFrame(l)
print(df)
# df.shape
# df.size
# df.ndim
```

```python
#create data frame from nested list

l=[["apple",102],["orange",48],["grapes",56]]

df=pd.DataFrame(l,columns=["fruits","price"])
df
```

```python
#create a data frame with id name and location and add 4 rows

l=[[1,"farhan","tvm"],[2,"manu","kochi"],[3,"anu","tvm"],[4,"abu","calicut"]]

df=pd.DataFrame(l,columns=["id","name","location"])
df

print(df.tail(1))
df.head(1)
```
